### Clean HTML from Email Text

Remove HTML tags and extract human-readable content from email bodies.

```python
from bs4 import BeautifulSoup
import re

def clean_email_html(html_text: str | None) -> str:
    """
    Extract clean, human-readable text from HTML email content.
    
    Removes:
    - All HTML tags
    - Script and style elements
    - Excessive whitespace and newlines
    - Email headers (From:, To:, Sent:, etc.)
    - Common email signatures and disclaimers
    
    Args:
        html_text: Raw HTML email content
        
    Returns:
        Clean, human-readable text content
    """
    if pd.isna(html_text) or html_text == '':
        return ''
    
    try:
        # Parse HTML with BeautifulSoup
        soup = BeautifulSoup(str(html_text), 'html.parser')
        
        # Remove script, style, and meta tags (non-content elements)
        for element in soup(['script', 'style', 'meta', 'link', 'head']):
            element.decompose()
        
        # Get text content
        text = soup.get_text(separator=' ', strip=True)
        
        # Clean up common email artifacts
        # Remove email headers (From:, To:, Sent:, Subject:, etc.)
        text = re.sub(r'^(From|To|Sent|Date|Subject|Cc|Bcc):\s*.*?$', '', text, flags=re.MULTILINE | re.IGNORECASE)
        
        # Remove common disclaimer patterns
        disclaimer_patterns = [
            r'This email.*?confidential.*?intended recipient.*?[\r\n]+',
            r'CONFIDENTIALITY NOTICE:.*?[\r\n]+',
            r'This message.*?confidential.*?[\r\n]+',
            r'If you are not the intended recipient.*?[\r\n]+',
            r'Please consider the environment before printing this email.*?[\r\n]+',
        ]
        for pattern in disclaimer_patterns:
            text = re.sub(pattern, '', text, flags=re.IGNORECASE | re.DOTALL)
        
        # Remove excessive whitespace
        text = re.sub(r'\s+', ' ', text)  # Multiple spaces to single space
        text = re.sub(r'\n\s*\n+', '\n\n', text)  # Multiple newlines to double newline
        
        # Remove leading/trailing whitespace
        text = text.strip()
        
        return text
    
    except Exception as e:
        # If parsing fails, return original text with minimal cleaning
        print(f"Warning: HTML parsing failed, returning minimally cleaned text. Error: {e}")
        # At minimum, remove HTML tags with regex
        text = re.sub(r'<[^>]+>', ' ', str(html_text))
        text = re.sub(r'\s+', ' ', text)
        return text.strip()

print("✅ HTML cleaning function created: clean_email_html()")
```

### Optimized HTML Cleaning (Faster Single-Process Version)

A faster version that uses lxml directly and vectorized operations.

```python
import re
from bs4 import BeautifulSoup

# Pre-compile regex patterns for better performance
HEADER_PATTERN = re.compile(r'^(From|To|Sent|Date|Subject|Cc|Bcc):\s*.*?$', re.MULTILINE | re.IGNORECASE)
DISCLAIMER_PATTERNS = [
    re.compile(r'This email.*?confidential.*?intended recipient.*?[\r\n]+', re.IGNORECASE | re.DOTALL),
    re.compile(r'CONFIDENTIALITY NOTICE:.*?[\r\n]+', re.IGNORECASE | re.DOTALL),
    re.compile(r'This message.*?confidential.*?[\r\n]+', re.IGNORECASE | re.DOTALL),
    re.compile(r'If you are not the intended recipient.*?[\r\n]+', re.IGNORECASE | re.DOTALL),
    re.compile(r'Please consider the environment before printing this email.*?[\r\n]+', re.IGNORECASE | re.DOTALL),
]
WHITESPACE_PATTERN = re.compile(r'\s+')
NEWLINE_PATTERN = re.compile(r'\n\s*\n+')
HTML_TAG_PATTERN = re.compile(r'<[^>]+>')

def clean_email_html_optimized(html_text: str | None) -> str:
    """
    Optimized version of clean_email_html with pre-compiled regex patterns.
    
    Args:
        html_text: Raw HTML email content
        
    Returns:
        Clean, human-readable text content
    """
    if pd.isna(html_text) or html_text == '':
        return ''
    
    try:
        # Parse HTML with BeautifulSoup (lxml parser is fastest)
        soup = BeautifulSoup(str(html_text), 'lxml')
        
        # Remove non-content elements
        for element in soup(['script', 'style', 'meta', 'link', 'head']):
            element.decompose()
        
        # Get text content
        text = soup.get_text(separator=' ', strip=True)
        
        # Remove email headers
        text = HEADER_PATTERN.sub('', text)
        
        # Remove disclaimers
        for pattern in DISCLAIMER_PATTERNS:
            text = pattern.sub('', text)
        
        # Clean whitespace
        text = WHITESPACE_PATTERN.sub(' ', text)
        text = NEWLINE_PATTERN.sub('\n\n', text)
        
        return text.strip()
    
    except Exception:
        # Fallback: minimal cleaning with regex
        text = HTML_TAG_PATTERN.sub(' ', str(html_text))
        text = WHITESPACE_PATTERN.sub(' ', text)
        return text.strip()

print("✅ Optimized cleaning function created: clean_email_html_optimized()")
print("   - Uses pre-compiled regex patterns for ~2-3x speed improvement")
print("   - Faster lxml parser")
```

### Parallel HTML Cleaning Function (Optional)

Only needed if you want to use parallel processing for very large datasets (50k+ emails).

```python
from multiprocessing import Pool, cpu_count
import time

def clean_email_batch(batch_data):
    """
    Clean a batch of emails (for parallel processing).
    
    Args:
        batch_data: Tuple of (batch_index, email_texts_list)
        
    Returns:
        Tuple of (batch_index, cleaned_texts_list)
    """
    batch_idx, emails = batch_data
    cleaned = [clean_email_html_optimized(email) for email in emails]
    return batch_idx, cleaned


def parallel_clean_emails(df, column_name='EMAIL_TEXT', num_workers=None, chunk_size=1000):
    """
    Clean HTML from emails using parallel processing.
    
    Args:
        df: DataFrame with email data
        column_name: Name of column containing HTML email text
        num_workers: Number of parallel workers (default: CPU count - 1)
        chunk_size: Number of emails per batch
        
    Returns:
        List of cleaned email texts in same order as input
    """
    if column_name not in df.columns:
        print(f"⚠️ Column '{column_name}' not found in dataframe")
        return []
    
    # Set number of workers
    if num_workers is None:
        num_workers = max(1, cpu_count() - 1)  # Leave one core free
    
    total_emails = len(df)
    print(f"\n{'='*80}")
    print(f"🚀 PARALLEL HTML CLEANING")
    print(f"{'='*80}")
    print(f"Total emails: {total_emails:,}")
    print(f"Workers: {num_workers}")
    print(f"Chunk size: {chunk_size:,}")
    print(f"Total chunks: {(total_emails + chunk_size - 1) // chunk_size}")
    print(f"{'='*80}\n")
    
    # Split data into chunks
    email_series = df[column_name]
    chunks = []
    for i in range(0, total_emails, chunk_size):
        chunk = email_series.iloc[i:i+chunk_size].tolist()
        chunks.append((i // chunk_size, chunk))
    
    print(f"Starting parallel processing...")
    start_time = time.time()
    
    # Process chunks in parallel
    with Pool(processes=num_workers) as pool:
        results = pool.map(clean_email_batch, chunks)
    
    end_time = time.time()
    elapsed_time = end_time - start_time
    
    # Sort results by batch index and flatten
    results.sort(key=lambda x: x[0])
    cleaned_emails = []
    for _, batch_cleaned in results:
        cleaned_emails.extend(batch_cleaned)
    
    print(f"\n✅ Parallel processing complete!")
    print(f"{'='*80}")
    print(f"⏱️  Total time: {elapsed_time:.1f} seconds ({elapsed_time/60:.1f} minutes)")
    print(f"   Average: {elapsed_time/total_emails*1000:.2f} ms per email")
    print(f"   Throughput: {total_emails/elapsed_time:.0f} emails/second")
    print(f"{'='*80}\n")
    
    return cleaned_emails

print("✅ Parallel cleaning function created: parallel_clean_emails()")
print("   Note: Uses clean_email_html_optimized() internally")
```

## 3. Extract Load Identifiers from Email Subjects

Create functions to:
1. Extract potential load identifiers from email subject lines (patterns like "Load #: 123456", "REF #: 123456", etc.)
2. Match extracted identifiers against the crosswalk data to find associated party codes

```python
import re
from typing import List

def extract_load_identifiers(subject: str) -> List[str]:
    """
    Extract potential load identifiers from an email subject line.
    
    Patterns detected:
    - 8-12 digit numbers (common load number format)
    - Letter-digit combinations (e.g., CHR123456, ABC-123456)
    - "Load #: 123456" or "Load#: 123456"
    - "REF #: 123456" or "REF#: 123456"
    - "Load 123456" or "Load# 123456"
    - "Reference: 123456" or "Ref: 123456"
    - "Order: 123456" or "ORD# 123456"
    - "Shipment: 123456" or "SHIP# 123456"
    - Standalone numbers that could be load numbers (6+ digits)
    - Alphanumeric identifiers like "CHR123456", "L-123456", etc.
    
    Args:
        subject: Email subject line
        
    Returns:
        List of unique potential load identifiers (as strings)
    """
    if pd.isna(subject) or not isinstance(subject, str):
        return []
    
    subject_upper = str(subject).upper()
    identifiers = set()
    
    # Pattern 1: 8-12 digit numbers (common load number format)
    pattern1 = re.findall(r'\b\d{8,12}\b', subject_upper)
    identifiers.update(pattern1)
    
    # Pattern 2: Letter-digit combinations (e.g., CHR123456, ABC-123456)
    pattern2 = re.findall(r'\b[A-Z]{2,4}-?\d{6,10}\b', subject_upper)
    identifiers.update(pattern2)
    
    # Pattern 3: Reference patterns (REF: 123456, REF# 123456, Reference: 123456)
    pattern3 = re.findall(r'(?:REF(?:ERENCE)?)[:\s#]*([A-Z0-9]{6,12})', subject_upper)
    identifiers.update(pattern3)
    
    # Pattern 4: Load patterns (LOAD: 123456, LID: 123456, Load# 123456)
    pattern4 = re.findall(r'(?:LOAD|LID)[:\s#]*([A-Z0-9]{6,12})', subject_upper)
    identifiers.update(pattern4)
    
    # Pattern 5: Order patterns (ORDER: 123456, ORD# 123456)
    pattern5 = re.findall(r'(?:ORDER|ORD)[:\s#]*([A-Z0-9]{6,12})', subject_upper)
    identifiers.update(pattern5)
    
    # Pattern 6: Shipment patterns (SHIPMENT: 123456, SHIP# 123456)
    pattern6 = re.findall(r'(?:SHIPMENT|SHIP)[:\s#]*([A-Z0-9]{6,12})', subject_upper)
    identifiers.update(pattern6)
    
    # Pattern 7: Common prefixes followed by numbers/alphanumeric
    # Matches: "CHR123456", "L-123456", "LOAD-123"
    pattern7 = re.findall(r'\b(?:CHR|L)[:\-\s]?([A-Z0-9]+)\b', subject_upper)
    identifiers.update(pattern7)
    
    # Pattern 8: Standalone numbers that could be load numbers (6-7 digits)
    # This catches shorter numbers that might be load IDs without keywords
    pattern8 = re.findall(r'\b(\d{6,7})\b', subject_upper)
    identifiers.update(pattern8)
    
    # Pattern 9: Hash/pound sign followed by number
    # Matches: "#123456", "# 123456"
    pattern9 = re.findall(r'#\s*(\d{6,12})\b', subject_upper)
    identifiers.update(pattern9)
    
    # Clean up and return unique identifiers
    # Remove any empty strings and ensure all are uppercase
    cleaned_identifiers = [id.strip().upper() for id in identifiers if id.strip()]
    
    return list(set(cleaned_identifiers))  # Return unique values only


print("✓ Load identifier extraction function created")
print("  - extract_load_identifiers(): Extract IDs from subject line")
```

```python
# Batch processing function to enrich dataframe with crosswalk data
def enrich_emails_with_crosswalk(email_df, crosswalk_df, chunk_size=100000):
    """
    Process emails in chunks and add loadnumber, partycode, carrier_code, and carrier_name columns from crosswalk matches.
    
    Args:
        email_df: DataFrame with email data including SUBJECT column
        crosswalk_df: DataFrame with LOADNUMBER, PARTYCODE, CARRIER_CODE, CARRIER_NAME, REFNUMS (array) columns
        chunk_size: Number of rows to process per chunk
        
    Returns:
        DataFrame with added columns: extracted_identifiers, matched_refnums, loadnumbers, partycodes, carrier_code, carrier_name, match_count
    """
    import gc
    
    total_rows = len(email_df)
    num_chunks = (total_rows + chunk_size - 1) // chunk_size
    
    print(f"\n{'='*80}")
    print(f"Enriching Email DataFrame with Crosswalk Data (Batch Processing)")
    print(f"{'='*80}")
    print(f"Total emails: {total_rows:,}")
    print(f"Chunk size: {chunk_size:,}")
    print(f"Number of chunks: {num_chunks}")
    print(f"Crosswalk records: {len(crosswalk_df):,}")
    print(f"{'='*80}\n")
    
    # Build crosswalk lookup once for efficiency (optimized vectorized approach)
    # New structure: REFNUM -> load/party/carrier info
    print("Building crosswalk lookup dictionary (optimized)...")
    crosswalk_lookup = {}
    
    # Use vectorized operations instead of iterrows for better performance
    for idx in range(len(crosswalk_df)):
        row = crosswalk_df.iloc[idx]
        refnums_list = row['REFNUMS']
        
        if isinstance(refnums_list, list) and len(refnums_list) > 0:
            # Pre-create the lookup value once for all REFNUMs in this row
            lookup_value = {
                'loadnumber': str(row['LOADNUMBER']),
                'partycode': row['PARTYCODE'],
                'carrier_code': row['CARRIER_CODE'] if pd.notna(row.get('CARRIER_CODE')) else None,
                'carrier_name': row['CARRIER_NAME'] if pd.notna(row.get('CARRIER_NAME')) else None
            }
            
            # Map each REFNUM to the same lookup value
            for refnum in refnums_list:
                if pd.notna(refnum):
                    refnum_upper = str(refnum).upper()
                    if refnum_upper:
                        crosswalk_lookup[refnum_upper] = lookup_value
    
    print(f"✓ Crosswalk lookup built with {len(crosswalk_lookup):,} unique REFNUMs\n")
    
    # Initialize result lists
    all_extracted_ids = []
    all_matched_refnums = []
    all_loadnumbers = []
    all_partycodes = []
    all_carrier_codes = []
    all_carrier_names = []
    all_match_counts = []
    
    # Process in chunks
    for chunk_num in range(num_chunks):
        start_idx = chunk_num * chunk_size
        end_idx = min((chunk_num + 1) * chunk_size, total_rows)
        
        print(f"📧 Processing chunk {chunk_num + 1}/{num_chunks}: rows {start_idx:,} to {end_idx:,}...")
        
        # Get chunk
        chunk = email_df.iloc[start_idx:end_idx]
        
        # Process each email in chunk
        chunk_extracted = []
        chunk_matched = []
        chunk_loadnums = []
        chunk_partycodes = []
        chunk_carrier_codes = []
        chunk_carrier_names = []
        chunk_counts = []
        
        for _, row in chunk.iterrows():
            # Extract identifiers
            extracted_ids = extract_load_identifiers(row['SUBJECT'])
            
            # Match against crosswalk
            matched_refnums = []
            loadnumbers = set()
            partycodes = set()
            carrier_code = None  # Single value, not a set
            carrier_name = None  # Single value, not a set
            
            for identifier in extracted_ids:
                identifier_upper = identifier.upper()
                if identifier_upper in crosswalk_lookup:
                    matched_refnums.append(identifier_upper)
                    loadnumbers.add(crosswalk_lookup[identifier_upper]['loadnumber'])
                    partycodes.add(crosswalk_lookup[identifier_upper]['partycode'])
                    # Get the first carrier_code and carrier_name found (since there's only one per email)
                    if carrier_code is None:
                        carrier_code = crosswalk_lookup[identifier_upper]['carrier_code']
                    if carrier_name is None:
                        carrier_name = crosswalk_lookup[identifier_upper]['carrier_name']
            
            # Store results
            chunk_extracted.append(extracted_ids)
            chunk_matched.append(matched_refnums)
            chunk_loadnums.append(list(loadnumbers))
            chunk_partycodes.append(list(partycodes))
            chunk_carrier_codes.append(carrier_code)  # Store as single value
            chunk_carrier_names.append(carrier_name)  # Store as single value
            chunk_counts.append(len(matched_refnums))
        
        # Add chunk results to all results
        all_extracted_ids.extend(chunk_extracted)
        all_matched_refnums.extend(chunk_matched)
        all_loadnumbers.extend(chunk_loadnums)
        all_partycodes.extend(chunk_partycodes)
        all_carrier_codes.extend(chunk_carrier_codes)
        all_carrier_names.extend(chunk_carrier_names)
        all_match_counts.extend(chunk_counts)
        
        # Clean up
        del chunk
        gc.collect()
    
    # Add columns to original dataframe
    print(f"\n✓ All chunks processed. Adding columns to dataframe...")
    email_df['extracted_identifiers'] = all_extracted_ids
    email_df['matched_refnums'] = all_matched_refnums
    email_df['loadnumbers'] = all_loadnumbers
    email_df['partycodes'] = all_partycodes
    email_df['carrier_code'] = all_carrier_codes  # Single string value
    email_df['carrier_name'] = all_carrier_names  # Single string value
    email_df['match_count'] = all_match_counts
    
    # Print summary statistics
    print(f"\n{'='*80}")
    print(f"📊 Enrichment Summary")
    print(f"{'='*80}")
    
    emails_with_extracted = sum(1 for x in all_extracted_ids if len(x) > 0)
    emails_with_matches = sum(1 for x in all_match_counts if x > 0)
    total_loadnums = sum(len(x) for x in all_loadnumbers)
    total_partycodes = sum(len(x) for x in all_partycodes)
    emails_with_carrier_code = sum(1 for x in all_carrier_codes if pd.notna(x))
    
    print(f"Emails with extracted identifiers: {emails_with_extracted:,} ({emails_with_extracted/total_rows*100:.1f}%)")
    print(f"Emails with crosswalk matches: {emails_with_matches:,} ({emails_with_matches/total_rows*100:.1f}%)")
    print(f"Total load numbers found: {total_loadnums:,}")
    print(f"Total party codes found: {total_partycodes:,}")
    print(f"Emails with carrier code: {emails_with_carrier_code:,} ({emails_with_carrier_code/total_rows*100:.1f}%)")
    print(f"Average identifiers per email: {sum(len(x) for x in all_extracted_ids)/total_rows:.2f}")
    print(f"Average matches per email: {sum(all_match_counts)/total_rows:.2f}")
    print(f"{'='*80}\n")
    
    return email_df

print("✅ Batch enrichment function created: enrich_emails_with_crosswalk()")
```

```python
def extract_load_identifiers_from_body(body_text: str, limit: int = 20) -> List[str]:
    """
    Extract potential load identifiers from cleaned email body text.
    
    Uses same patterns as subject line extraction but applies to email body.
    Limits results to avoid extracting too many false positives from long emails.
    
    Args:
        body_text: Cleaned email body text
        limit: Maximum number of identifiers to extract (default: 20)
        
    Returns:
        List of unique potential load identifiers (as strings)
    """
    if pd.isna(body_text) or body_text == '':
        return []
    
    body_upper = str(body_text).upper()
    identifiers = set()
    
    # Pattern 1: 8-12 digit numbers (common load number format)
    pattern1 = re.findall(r'\b\d{8,12}\b', body_upper)
    identifiers.update(pattern1)
    
    # Pattern 2: Letter-digit combinations (e.g., CHR123456, ABC-123456)
    pattern2 = re.findall(r'\b[A-Z]{2,4}-?\d{6,10}\b', body_upper)
    identifiers.update(pattern2)
    
    # Pattern 3: Reference patterns (REF: 123456, REF# 123456, Reference: 123456)
    pattern3 = re.findall(r'(?:REF(?:ERENCE)?)[:\s#]*([A-Z0-9]{6,12})', body_upper)
    identifiers.update(pattern3)
    
    # Pattern 4: Load patterns (LOAD: 123456, LID: 123456, Load# 123456)
    pattern4 = re.findall(r'(?:LOAD|LID)[:\s#]*([A-Z0-9]{6,12})', body_upper)
    identifiers.update(pattern4)
    
    # Pattern 5: Order patterns (ORDER: 123456, ORD# 123456)
    pattern5 = re.findall(r'(?:ORDER|ORD)[:\s#]*([A-Z0-9]{6,12})', body_upper)
    identifiers.update(pattern5)
    
    # Pattern 6: Shipment patterns (SHIPMENT: 123456, SHIP# 123456)
    pattern6 = re.findall(r'(?:SHIPMENT|SHIP)[:\s#]*([A-Z0-9]{6,12})', body_upper)
    identifiers.update(pattern6)
    
    # Pattern 7: Common prefixes followed by numbers/alphanumeric
    pattern7 = re.findall(r'\b(?:CHR|L)[:\-\s]?([A-Z0-9]+)\b', body_upper)
    identifiers.update(pattern7)
    
    # Pattern 8: Hash/pound sign followed by number
    pattern8 = re.findall(r'#\s*(\d{6,12})\b', body_upper)
    identifiers.update(pattern8)
    
    # NOTE: Deliberately excluding 6-7 digit standalone numbers from body text
    # to reduce false positives (dates, phone numbers, etc.)
    
    # Clean up and return unique identifiers
    cleaned_identifiers = [id.strip().upper() for id in identifiers if id.strip()]
    
    # Filter out invalid identifiers:
    # 1. All-string identifiers (no digits) - e.g., "ABC", "LOAD"
    # 2. Single character identifiers - e.g., "A", "1"
    filtered_identifiers = []
    for identifier in cleaned_identifiers:
        # Skip single character identifiers
        if len(identifier) <= 1:
            continue
        # Skip all-letter identifiers (must contain at least one digit)
        if identifier.isalpha():
            continue
        # Keep valid identifiers
        filtered_identifiers.append(identifier)
    
    unique_identifiers = list(set(filtered_identifiers))
    
    # Limit results to avoid extracting too many from long emails
    return unique_identifiers[:limit]


print("✓ Body text identifier extraction function created")
print("  - extract_load_identifiers_from_body(): Extract IDs from email body")
print("  - Excludes: single characters, all-letter identifiers")
print("  - Note: Limits to 20 identifiers per email to reduce false positives")
```

### Extract Load Identifiers from Email Body Text

Similar to subject line extraction, extract potential load identifiers from the cleaned email body text.

```python
def enrich_emails_with_crosswalk_enhanced(email_df, crosswalk_df, chunk_size=100000, 
                                          extract_from_body=True, body_column='EMAIL_TEXT_CLEAN'):
    """
    Process emails in chunks and extract identifiers from BOTH subject and body text.
    
    Args:
        email_df: DataFrame with email data including SUBJECT and body text columns
        crosswalk_df: DataFrame with LOADNUMBER, PARTYCODE, CARRIER_CODE, CARRIER_NAME, REFNUMS (array) columns
        chunk_size: Number of rows to process per chunk
        extract_from_body: Whether to also extract from email body (default: True)
        body_column: Column name containing cleaned email body text (default: 'EMAIL_TEXT_CLEAN')
        
    Returns:
        DataFrame with added columns including identifiers from subject AND body
    """
    import gc
    
    total_rows = len(email_df)
    num_chunks = (total_rows + chunk_size - 1) // chunk_size
    
    print(f"\n{'='*80}")
    print(f"Enhanced Email Enrichment (Subject + Body Text Extraction)")
    print(f"{'='*80}")
    print(f"Total emails: {total_rows:,}")
    print(f"Chunk size: {chunk_size:,}")
    print(f"Number of chunks: {num_chunks}")
    print(f"Crosswalk records: {len(crosswalk_df):,}")
    print(f"Extract from body: {extract_from_body}")
    if extract_from_body:
        print(f"Body text column: {body_column}")
    print(f"{'='*80}\n")
    
    # Build crosswalk lookup once for efficiency (optimized vectorized approach)
    # New structure: REFNUM -> load/party/carrier info
    print("Building crosswalk lookup dictionary (optimized)...")
    crosswalk_lookup = {}
    
    # Use vectorized operations instead of iterrows for better performance
    for idx in range(len(crosswalk_df)):
        row = crosswalk_df.iloc[idx]
        refnums_list = row['REFNUMS']
        
        if isinstance(refnums_list, list) and len(refnums_list) > 0:
            # Pre-create the lookup value once for all REFNUMs in this row
            lookup_value = {
                'loadnumber': str(row['LOADNUMBER']),
                'partycode': row['PARTYCODE'],
                'carrier_code': row['CARRIER_CODE'] if pd.notna(row.get('CARRIER_CODE')) else None,
                'carrier_name': row['CARRIER_NAME'] if pd.notna(row.get('CARRIER_NAME')) else None
            }
            
            # Map each REFNUM to the same lookup value
            for refnum in refnums_list:
                if pd.notna(refnum):
                    refnum_upper = str(refnum).upper()
                    if refnum_upper:
                        crosswalk_lookup[refnum_upper] = lookup_value
    
    print(f"✓ Crosswalk lookup built with {len(crosswalk_lookup):,} unique REFNUMs\n")
    
    # Initialize result lists
    all_extracted_ids_subject = []
    all_extracted_ids_body = []
    all_extracted_ids_combined = []
    all_matched_refnums = []
    all_loadnumbers = []
    all_partycodes = []
    all_carrier_codes = []
    all_carrier_names = []
    all_match_counts = []
    all_match_source = []  # Track where match came from: 'subject', 'body', or 'both'
    
    # Process in chunks
    for chunk_num in range(num_chunks):
        start_idx = chunk_num * chunk_size
        end_idx = min((chunk_num + 1) * chunk_size, total_rows)
        
        print(f"📧 Processing chunk {chunk_num + 1}/{num_chunks}: rows {start_idx:,} to {end_idx:,}...")
        
        # Get chunk
        chunk = email_df.iloc[start_idx:end_idx]
        
        # Process each email in chunk
        chunk_extracted_subject = []
        chunk_extracted_body = []
        chunk_extracted_combined = []
        chunk_matched = []
        chunk_loadnums = []
        chunk_partycodes = []
        chunk_carrier_codes = []
        chunk_carrier_names = []
        chunk_counts = []
        chunk_sources = []
        
        for _, row in chunk.iterrows():
            # Extract identifiers from subject
            extracted_ids_subject = extract_load_identifiers(row['SUBJECT'])
            
            # Extract identifiers from body (if enabled)
            extracted_ids_body = []
            if extract_from_body and body_column in email_df.columns:
                extracted_ids_body = extract_load_identifiers_from_body(row[body_column])
            
            # Combine unique identifiers from both sources
            all_ids = list(set(extracted_ids_subject + extracted_ids_body))
            
            # Match against crosswalk
            matched_refnums = []
            matched_from_subject = []
            matched_from_body = []
            loadnumbers = set()
            partycodes = set()
            carrier_code = None
            carrier_name = None
            
            for identifier in all_ids:
                identifier_upper = identifier.upper()
                if identifier_upper in crosswalk_lookup:
                    matched_refnums.append(identifier_upper)
                    loadnumbers.add(crosswalk_lookup[identifier_upper]['loadnumber'])
                    partycodes.add(crosswalk_lookup[identifier_upper]['partycode'])
                    
                    # Track source of match
                    if identifier_upper in [x.upper() for x in extracted_ids_subject]:
                        matched_from_subject.append(identifier_upper)
                    if identifier_upper in [x.upper() for x in extracted_ids_body]:
                        matched_from_body.append(identifier_upper)
                    
                    # Get the first carrier_code and carrier_name found
                    if carrier_code is None:
                        carrier_code = crosswalk_lookup[identifier_upper]['carrier_code']
                    if carrier_name is None:
                        carrier_name = crosswalk_lookup[identifier_upper]['carrier_name']
            
            # Determine match source
            if len(matched_from_subject) > 0 and len(matched_from_body) > 0:
                match_source = 'both'
            elif len(matched_from_subject) > 0:
                match_source = 'subject'
            elif len(matched_from_body) > 0:
                match_source = 'body'
            else:
                match_source = None
            
            # Store results
            chunk_extracted_subject.append(extracted_ids_subject)
            chunk_extracted_body.append(extracted_ids_body)
            chunk_extracted_combined.append(all_ids)
            chunk_matched.append(matched_refnums)
            chunk_loadnums.append(list(loadnumbers))
            chunk_partycodes.append(list(partycodes))
            chunk_carrier_codes.append(carrier_code)
            chunk_carrier_names.append(carrier_name)
            chunk_counts.append(len(matched_refnums))
            chunk_sources.append(match_source)
        
        # Add chunk results to all results
        all_extracted_ids_subject.extend(chunk_extracted_subject)
        all_extracted_ids_body.extend(chunk_extracted_body)
        all_extracted_ids_combined.extend(chunk_extracted_combined)
        all_matched_refnums.extend(chunk_matched)
        all_loadnumbers.extend(chunk_loadnums)
        all_partycodes.extend(chunk_partycodes)
        all_carrier_codes.extend(chunk_carrier_codes)
        all_carrier_names.extend(chunk_carrier_names)
        all_match_counts.extend(chunk_counts)
        all_match_source.extend(chunk_sources)
        
        # Clean up
        del chunk
        gc.collect()
    
    # Add columns to original dataframe
    print(f"\n✓ All chunks processed. Adding columns to dataframe...")
    email_df['extracted_ids_subject'] = all_extracted_ids_subject
    email_df['extracted_ids_body'] = all_extracted_ids_body
    email_df['extracted_identifiers'] = all_extracted_ids_combined
    email_df['matched_refnums'] = all_matched_refnums
    email_df['loadnumbers'] = all_loadnumbers
    email_df['partycodes'] = all_partycodes
    email_df['carrier_code'] = all_carrier_codes
    email_df['carrier_name'] = all_carrier_names
    email_df['match_count'] = all_match_counts
    email_df['match_source'] = all_match_source
    
    # Print summary statistics
    print(f"\n{'='*80}")
    print(f"📊 Enhanced Enrichment Summary")
    print(f"{'='*80}")
    
    emails_with_extracted_subject = sum(1 for x in all_extracted_ids_subject if len(x) > 0)
    emails_with_extracted_body = sum(1 for x in all_extracted_ids_body if len(x) > 0)
    emails_with_extracted_any = sum(1 for x in all_extracted_ids_combined if len(x) > 0)
    emails_with_matches = sum(1 for x in all_match_counts if x > 0)
    
    # Count matches by source
    matches_from_subject = sum(1 for x in all_match_source if x == 'subject')
    matches_from_body = sum(1 for x in all_match_source if x == 'body')
    matches_from_both = sum(1 for x in all_match_source if x == 'both')
    
    total_loadnums = sum(len(x) for x in all_loadnumbers)
    total_partycodes = sum(len(x) for x in all_partycodes)
    emails_with_carrier_code = sum(1 for x in all_carrier_codes if pd.notna(x))
    
    print(f"\n📧 Identifier Extraction:")
    print(f"  From subject only: {emails_with_extracted_subject:,} emails")
    print(f"  From body only: {emails_with_extracted_body:,} emails")
    print(f"  From either source: {emails_with_extracted_any:,} emails ({emails_with_extracted_any/total_rows*100:.1f}%)")
    
    print(f"\n🎯 Crosswalk Matches:")
    print(f"  Total emails with matches: {emails_with_matches:,} ({emails_with_matches/total_rows*100:.1f}%)")
    print(f"  Matched from subject: {matches_from_subject:,}")
    print(f"  Matched from body: {matches_from_body:,}")
    print(f"  Matched from both: {matches_from_both:,}")
    
    print(f"\n📦 Results:")
    print(f"  Total load numbers found: {total_loadnums:,}")
    print(f"  Total party codes found: {total_partycodes:,}")
    print(f"  Emails with carrier code: {emails_with_carrier_code:,} ({emails_with_carrier_code/total_rows*100:.1f}%)")
    print(f"  Average identifiers per email: {sum(len(x) for x in all_extracted_ids_combined)/total_rows:.2f}")
    print(f"  Average matches per email: {sum(all_match_counts)/total_rows:.2f}")
    print(f"{'='*80}\n")
    
    return email_df

print("✅ Enhanced enrichment function created: enrich_emails_with_crosswalk_enhanced()")
print("   - Extracts from both SUBJECT and EMAIL_TEXT_CLEAN")
print("   - Tracks match source (subject/body/both)")
```

## 4. Filter Out Internal CHR Emails

Remove emails where CHR is sending only to other CHR recipients (internal emails).
Keep emails where:
- SENDER_IS_CHR = 1 AND at least one recipient is external (RECIPIENT_IS_CHR contains 0)
- OR SENDER_IS_CHR = 0 (external sender)

```python
def filter_external_emails(df):
    """
    Filter out internal CHR emails (CHR sender to only CHR recipients).
    
    Keeps emails where:
    - SENDER_IS_CHR = 1 AND RECIPIENT_IS_CHR contains at least one 0 (external recipient)
    - OR SENDER_IS_CHR = 0 (external sender - keep all regardless of recipients)
    
    Removes emails where:
    - SENDER_IS_CHR = 1 AND RECIPIENT_IS_CHR = [1] (only internal CHR recipients)
    
    Args:
        df: DataFrame with SENDER_IS_CHR and RECIPIENT_IS_CHR columns
        
    Returns:
        Filtered DataFrame with internal CHR emails removed
    """
    original_count = len(df)
    
    # Create a mask for rows to keep
    # Keep if: sender is external (SENDER_IS_CHR = 0) OR (sender is CHR AND has external recipients)
    def has_external_recipient(recipient_array):
        """Check if the RECIPIENT_IS_CHR array contains at least one 0."""
        if pd.isna(recipient_array):
            return False
        if isinstance(recipient_array, list):
            return 0 in recipient_array
        return False
    
    # Mask: Keep all emails where sender is NOT CHR (0), 
    # OR where sender IS CHR (1) but has at least one external recipient (0 in array)
    mask = (
        (df['SENDER_IS_CHR'] == 0) |  # External sender - keep all
        (
            (df['SENDER_IS_CHR'] == 1) &  # CHR sender
            df['RECIPIENT_IS_CHR'].apply(has_external_recipient)  # Has external recipient
        )
    )
    
    df_filtered = df[mask].copy()
    
    removed_count = original_count - len(df_filtered)
    
    print("="*80)
    print("📧 INTERNAL EMAIL FILTERING")
    print("="*80)
    print(f"Original email count: {original_count:,}")
    print(f"Emails removed (CHR → only CHR recipients): {removed_count:,}")
    print(f"Emails remaining: {len(df_filtered):,}")
    print(f"Retention rate: {len(df_filtered)/original_count*100:.2f}%")
    print("="*80)
    
    return df_filtered

print("✅ Filter function created: filter_external_emails()")
```

## 5. Analyze Non-CHR Email Recipients by Carrier

Analyze which external email addresses each carrier communicates with and the distribution of emails sent to each recipient.

```python
def analyze_carrier_recipients(df):
    """
    Analyze non-CHR email recipients by carrier code.
    
    For each carrier, this function:
    1. Extracts all recipient email addresses from the RECIPIENT field
    2. Filters to only non-CHR emails (not ending in @chrobinson.com)
    3. Calculates the percentage of emails sent to each non-CHR recipient
    
    Args:
        df: DataFrame with CARRIER_CODE and RECIPIENT columns
        
    Returns:
        DataFrame with columns:
        - carrier_code: The carrier code
        - non_chr_recipients: Array of non-CHR email addresses
        - recipient_percentages: Array of percentages (corresponding to non_chr_recipients)
    """
    import re
    from collections import Counter
    
    def extract_non_chr_emails(recipient_field):
        """Extract non-CHR email addresses from the RECIPIENT field."""
        if pd.isna(recipient_field) or recipient_field == '':
            return []
        
        # Extract all email addresses using regex
        email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
        emails = re.findall(email_pattern, str(recipient_field))
        
        # Filter to only non-CHR emails (not ending in @chrobinson.com)
        non_chr_emails = [
            email.lower() for email in emails 
            if not email.lower().endswith('@chrobinson.com')
        ]
        
        return non_chr_emails
    
    # Filter to only rows with carrier_code
    df_with_carrier = df[df['carrier_code'].notna()].copy()
    
    print("="*80)
    print("📊 ANALYZING NON-CHR RECIPIENTS BY CARRIER")
    print("="*80)
    print(f"Total emails with carrier code: {len(df_with_carrier):,}")
    print(f"Processing recipient email extraction...")
    print("="*80 + "\n")
    
    # Group by carrier_code and analyze recipients
    results = []
    
    for carrier_code, group in df_with_carrier.groupby('carrier_code'):
        # Get carrier_name (should be the same for all rows in group)
        carrier_name = group['carrier_name'].iloc[0] if 'carrier_name' in group.columns else None
        
        # Count all non-CHR emails for this carrier
        all_non_chr_emails = []
        
        for _, row in group.iterrows():
            non_chr_emails = extract_non_chr_emails(row['RECIPIENT'])
            all_non_chr_emails.extend(non_chr_emails)
        
        if len(all_non_chr_emails) == 0:
            # No non-CHR recipients for this carrier
            continue
        
        # Count occurrences of each email
        email_counts = Counter(all_non_chr_emails)
        total_emails = len(all_non_chr_emails)
        
        # Sort by count (descending) to get most common recipients first
        sorted_emails = sorted(email_counts.items(), key=lambda x: x[1], reverse=True)
        
        # Extract emails and percentages in the same order
        non_chr_recipients = [email for email, count in sorted_emails]
        recipient_percentages = [
            round((count / total_emails) * 100, 2) 
            for email, count in sorted_emails
        ]
        
        results.append({
            'carrier_code': carrier_code,
            'carrier_name': carrier_name,
            'non_chr_recipients': non_chr_recipients,
            'recipient_percentages': recipient_percentages,
            'total_non_chr_emails': total_emails,
            'unique_recipients': len(non_chr_recipients)
        })
    
    # Create result dataframe
    result_df = pd.DataFrame(results)
    
    # Sort by total_non_chr_emails descending
    result_df = result_df.sort_values('total_non_chr_emails', ascending=False).reset_index(drop=True)
    
    print("\n" + "="*80)
    print("📈 ANALYSIS SUMMARY")
    print("="*80)
    print(f"Carriers with non-CHR recipients: {len(result_df):,}")
    print(f"Total non-CHR email instances analyzed: {result_df['total_non_chr_emails'].sum():,}")
    print(f"\nTop 5 carriers by non-CHR email volume:")
    for i, row in result_df.head().iterrows():
        carrier_display = f"{row['carrier_code']}"
        if pd.notna(row.get('carrier_name')):
            carrier_display += f" ({row['carrier_name']})"
        print(f"  {i+1}. {carrier_display}: {row['total_non_chr_emails']:,} emails to {row['unique_recipients']} unique recipients")
    print("="*80)
    
    return result_df

print("✅ Carrier recipient analysis function created: analyze_carrier_recipients()")
```

```python
# Analyze non-CHR recipients by carrier
carrier_recipients_df = analyze_carrier_recipients(df_final_filtered)

print(f"\n✅ Analysis complete!")
print(f"\nResult dataframe columns:")
print(f"  - carrier_code: Carrier identifier")
print(f"  - carrier_name: Carrier name")
print(f"  - non_chr_recipients: Array of non-CHR email addresses (sorted by frequency)")
print(f"  - recipient_percentages: Percentage of emails to each recipient (same order)")
print(f"  - total_non_chr_emails: Total number of non-CHR email instances")
print(f"  - unique_recipients: Count of unique non-CHR recipients")

carrier_recipients_df.head()
```
